```python
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("/Users/zahiramohammed/Desktop/DAPM_dataset_new.csv") #read the dataset
```

```python
# Data Cleaning
# -----No cleaning needed for this dataset(as it is previously cleaned data)------

# Data Integration
# -----Only one dataset is considered, so no integration is needed----

# Data Reduction 
# ------No reduction required, as there are considerable number of variables--------

# """"""Data Transformation""""""

# Normalize BMI using StandardScaler
scaler = StandardScaler()
bmi_scaled = scaler.fit_transform(data[['bmi']])
data['bmi_scaled'] = bmi_scaled
```

```python
# """"""Data Discretization""""""""""
# Discretize Age
def discretized_age(age):
    if age <= 18: 
        return 0
    elif age <=39:
        return 1
    elif age <= 64: 
        return 2
    else:
        return 3
data['age_discretized'] = data['age'].apply(discretized_age)

# Discretize Blood Glucose Level 
def discretized_glucose(col):  
    if col < 100:
        return 0 
    elif 100 <= col <= 125:
        return 1
    else:
        return 2

data['glucose_discretized'] = data['blood_glucose_level'].apply(discretized_glucose)
```
